```python
X_train, X_test, y_train, y_test = train_test_split(ffadata, y, test_size=0.2, random_state=42)
X_train = X_train.to_df()
X_test = X_test.to_df()
num_features = X_train.shape[1]
```

```python
X_train = np.expand_dims(X_train, axis=2)
X_test = np.expand_dims(X_test, axis=2)
```

```python
y_train = pd.DataFrame(to_categorical(y_train))
y_test = pd.DataFrame(to_categorical(y_test))
```

```python
num_features = X_train_reshape.shape[1]
print(num_features)
```

```python
model.trainable = True
```

```python
new_model = Sequential()
new_model.add(Conv1D(filters=32 ,kernel_size=1, activation='relu', input_shape=(num_features,1)))
new_model.add(MaxPooling1D(pool_size=1))
new_model.add(Conv1D(filters=64, kernel_size=1, activation='relu'))
new_model.add(MaxPooling1D(pool_size=1))
new_model.add(Conv1D(filters=128, kernel_size=1, activation='relu'))
for i, layer in enumerate(model.layers[1:len(new_model.layers)]):
    new_model.layers[i+1].set_weights(layer.get_weights())
```

```python
new_model.add(Dense(128, activation='relu'))
new_model.add(Dense(8, activation='softmax'))
```

```python
model.summary()
```

```python
new_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
# Train the model
new_model.fit(X_train_reshape, y_train, epochs=10, batch_size=32, validation_data=(X_test,y_test))
```
